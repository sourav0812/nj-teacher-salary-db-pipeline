## Question 3 (20 pts)

```python
#your code here

# we need to use this particualr database/schema 
mycursor.execute("USE nj_state_teachers_salaries")

# create table
##300 bc we want upto 300 characters input
mycursor.execute("""CREATE TABLE teachers_salaries (
id VARCHAR (300), 
last_name VARCHAR (300),
first_name VARCHAR (300),
county VARCHAR (300),
district VARCHAR (300),
school VARCHAR (300),
primary_job VARCHAR (300),
fte VARCHAR (300),
salary VARCHAR (300),
certificate VARCHAR (300),
subcategory VARCHAR (300),
teaching_route VARCHAR (300),
highly_qualified VARCHAR (300),
experience_district VARCHAR (300),
experience_nj VARCHAR (300),
experience_total VARCHAR (300) )

""" )

print("Table is created successfully") # able to see the table in workbench

```

**Using LOAD DATA statement (as discussed in Module 4 lectures) load the data from nj_teachers_salaries_pset3.csv to your table created in Question-3. Use of OPTIONALLY ENCLOSED BY  clause , TERMINATED BY clause and ESCAPED BY clause is recommended.** <br>

```python
#your code here
mycursor.execute("""LOAD DATA INFILE "/Users/souravchakraborty/cnf/nj_teachers_salaries_pset3.csv"
INTO TABLE teachers_salaries
FIELDS TERMINATED BY ','
OPTIONALLY ENCLOSED BY '"'
ESCAPED BY '\' 

LINES TERMINATED BY '\n'
IGNORE 1 ROWS
""")
mydb.commit() # to save changes to the database permenantly.

print("Data is loaded successfully")


```

## Question 5 (15 pts)

Run the cell below. The code checks if all the data rows were stored in the database. 

The code below assumes that you named your cursor object as mycursor(As specified in Question-1). If you named it differently, you can rename mycursor to match the variable name.



```python
cmd = "select count(*) from \
                 nj_state_teachers_salaries.teachers_salaries"
mycursor.execute(cmd)
count = mycursor.fetchone()[0]

print(f"Number of rows in teachers_salaries table : {count}")
```

## Question 6 (5 pts)

Run the cell below. The code checks if all the data columns were stored in the database. 

```python
cmd = """SELECT COUNT(*) \
                FROM INFORMATION_SCHEMA.COLUMNS \
                WHERE table_schema = 'nj_state_teachers_salaries' \
                AND table_name = 'teachers_salaries'"""
mycursor.execute(cmd)
count = mycursor.fetchone()[0]
print(f"Number of columns in teachers_salaries table : {count}")        
        
```
